```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_week1 = pd.read_csv(r'data/CAISO-demand-2021-05-week1.csv')
dfs = [df_week1]

for i in range(8, 32):
    if i < 10:
        df = pd.read_csv(r'data/CAISO-demand-2021050' + str(i) + '.csv')
    else:
        df = pd.read_csv(r'data/CAISO-demand-202105' + str(i) + '.csv')
    df.drop(axis=0, index=[0, 1, 3], inplace=True)
    df.iloc[0,0] = df.columns[0]
    df.rename(columns = {list(df)[0]:' '}, inplace=True)
    dfs.append(df)

# dfs[5].head()
# print(len(dfs))
df_month = pd.concat(dfs)

df_month.to_csv('CAISO-demand-2021-05.csv', index=False)
```
