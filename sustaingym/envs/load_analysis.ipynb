```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dfs = []

for i in range(1, 32):
    if i < 10:
        df = pd.read_csv('data/CAISO-netdemand-2021050%d.csv' % i, on_bad_lines='skip')
        label = 'Net Demand 05/0' + str(i) + '/2021'
    else:
        df = pd.read_csv('data/CAISO-netdemand-202105%d.csv' % i, on_bad_lines='skip')
        label = 'Net Demand 05/' + str(i) + '/2021'
    df.rename(columns={label: ' '}, inplace=True)
    df.iloc[1, 0] = label
    df.drop([0], inplace=True)
    dfs.append(df)

df_load = pd.concat(dfs)
df_load.to_csv('CAISO-netdemand-2021-05.csv')
df_load.head()
```
