```python
from io import StringIO
import pandas as pd
import requests

r = requests.get('https://www.caiso.com/outlook/SP/History/20190301/netdemand.csv')
df = pd.read_csv(StringIO(r.text))
df = df.T
label = 'Net Demand 03/01/2019'
df.rename(columns={label: ' '}, inplace=True)
df.rename(index={'Net demand': label}, inplace=True)
df = df.drop(['Hour ahead forecast', 'Current demand'])
df.head()
```

```python
from io import StringIO
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

r = requests.get('https://www.caiso.com/outlook/SP/History/20210401/netdemand.csv')
df = pd.read_csv(StringIO(r.text))
df = df.T
label = 'Net Demand 04/01/2021'
df.rename(columns={label: ' '}, inplace=True)
df.rename(index={'Net demand': label}, inplace=True)
df = df.drop(['Hour ahead forecast', 'Current demand'])

dfs = [df]

for i in tqdm(range(2, 31)):
    if i < 10:
        r = requests.get('https://www.caiso.com/outlook/SP/History/2021040' + str(i) + '/netdemand.csv')
        df = pd.read_csv(StringIO(r.text))
        df = df.T
        label = 'Net Demand 04/0' + str(i) + '/2021'
    else:
        r = requests.get('https://www.caiso.com/outlook/SP/History/202104' + str(i) + '/netdemand.csv')
        df = pd.read_csv(StringIO(r.text))
        df = df.T
        label = 'Net Demand 04/' + str(i) + '/2021'
    df.rename(columns={label: ' '}, inplace=True)
    df.rename(index={'Net demand': label}, inplace=True)
    df = df.drop(['Time', 'Hour ahead forecast', 'Current demand'])
    dfs.append(df)

df_load = pd.concat(dfs)
df_load.to_csv('CAISO-netdemand-2021-04.csv')
df_load.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dfs = []

for i in range(1, 32):
    if i < 10:
        df = pd.read_csv('data/CAISO-netdemand-2019030%d.csv' % i, on_bad_lines='skip')
        label = 'Net Demand 05/0' + str(i) + '/2021'
    else:
        df = pd.read_csv('data/CAISO-netdemand-201903%d.csv' % i, on_bad_lines='skip')
        label = 'Net Demand 05/' + str(i) + '/2021'
    df.rename(columns={label: ' '}, inplace=True)
    df.iloc[1, 0] = label
    df.drop([0], inplace=True)
    dfs.append(df)

df_load = pd.concat(dfs)
df_load.to_csv('CAISO-netdemand-2021-05.csv')
df_load.head()
```
